```python
import requests
from bs4 import BeautifulSoup
player_data = []

def getPageData(pageurl,player_data,max_col):
    row_data = []
    i=0
    res = requests.get(pageurl)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text,'html.parser')
    for nor in soup.select('td'):
        if max_col == 23 and i == 0:
            year = pageurl[-9:-5]
            year1 = str(int(year)+1)
            yearstr = year[-2:]+'-'+year1[-2:]
            row_data.append(yearstr)
        nor_string = str(nor)
        if 'col' + str(i) in nor_string:
            #nor_text = str(nor.text)
            #if nor_text == '' or nor_text == ' ':
                #row_data.append(0)
            #else:
            row_data.append(nor.text)
            i+=1
        if i>max_col:
            i=0
            player_data.append(row_data)
            row_data = []
    return player_data
```

```python
url = 'http://www.stat-nba.com/query.php?page={}&QueryType=ss&SsType=season&AT=avg&order=1&crtcol=season&Formular=leftbracketptsadd0.4multiplyfgsubstract0.7multiplyfgasubstract0.4multiplyleftbracketftasubstractftrightbracketadd0.7multiplyorbadd0.3multiplydrbaddstladd0.7multiplyastadd0.7multiplyblksubstract0.4multiplypfsubstracttovrightbracket&PageNum=200'
for i in range(154):
    pageurl = url.format(i)
    player_data = getPageData(pageurl,player_data,27)
```

```python
import pandas
df = pandas.DataFrame(player_data)
#df.replace('', 0, inplace=True)
df.tail(100)
```

```python
df.to_csv('player_data_with_missing.csv')
```
